```python
import pandas as pd

titanic = pd.read_csv("train.csv")

titanic
```

```python
titanic = titanic.loc[:,["Pclass", "Sex", "Age", "Survived"]]
titanic
```

```python
# 널값 제거
titanic = titanic.dropna(axis=0)
titanic
```

```python
# 정수화
titanic.loc[:,"Sex"].replace(["female","male"],[0,1],inplace=True)
titanic
```

```python
# 종속, 독립 변인 추출
titanic_input = titanic.loc[:,["Pclass", "Sex","Age"]]
titanic_target = titanic.loc[:,"Survived"]

print(titanic_input.count())
print(titanic_target.count())
```

```python
#넘파이로 변경
import numpy as np

tc_input_np = titanic_input.to_numpy()
tc_target_np = titanic_target.to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    tc_input_np, tc_target_np, random_state=42
)
```

```python
# 로지스틱 회귀 (분류)
# 경사하강법, 시그모이드, 자연상수, 로그

from sklearn.linear_model import LogisticRegression # 결정트리를 사용하는 것이 더 좋다.
lr = LogisticRegression(C=1, max_iter=1000) # C값이 높으면 규제를 약하게, C값이 작으면 규제를 강하게
lr.fit(train_input,train_target)
```

```python
print(lr.score(train_input,train_target))
print(lr.score(test_input,test_target))
```

### 계수를 좋게 하려면?
- 규제와 에포크
- 데이터가 더 많이 필요 (행이 필요함)
- 칼럼을 더 많이 선정해야함. (요금, 몇명에서 같이 왔는지)
- 더 좋은 모델을 선택해야함.
