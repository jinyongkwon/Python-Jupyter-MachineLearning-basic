# 판다스 Pandas
- 판다스는 R을 모티브로 하여 만든 파이썬 라이브러리이다.
- 판다스는 데이터분석을 위해 DataFrame을 사용하는데 열과 행으로 이루어진 표를 의미!!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = pd.Series([1,3,5, np.nan,6,8])
print(s)

s2 = pd.Series(['남','여','여','남'])
print(s2)
```

```python
datas = pd.date_range('20220517', periods=100) # 17일 부터 100개를 만듬
print(datas)
```

```python
df = pd.DataFrame({
    'A':1,
    'B':pd.Timestamp('20220517'),
    'C':pd.Series([1,2,3,4]),
    'D':'hello'
})

print(df.dtypes)
print(df)
```

```python
df.head(2) # 앞에서 부터 정한 갯수를 보여줌 = 디폴트는 5개
```

```python
df.tail(3) # 뒤에서 부터 정한갯수를 보여줌 = 디폴트 5
```

```python
df.index # 인덱스 확인
```

```python
df.columns # 칼럼 확인
```

```python
df.values # 값 확인
```

```python
df.describe() # 통계확인 => 숫자만 나옴 => 즉 정수화를 시켜야 확인가능
```

```python
# 칼럼과 index를 바꾼 데이터 프레임 ,RDB의 피벗 => 가로 세로를 바꿔줌
df.T 
```

```python
print(df)
df.sort_index(axis=1, ascending=False) # 1은 칼럼정렬, 0은 인덱스정렬 ,ascending=False는 내림차순정렬
```

```python
df.sort_index(axis=0, ascending=False) # 1은 칼럼정렬, 0은 인덱스정렬 ,ascending=False는 내림차순정렬
```

```python
df.sort_values(by='C', ascending=True) # 값으로 정렬 ,ascending=True는 오름차순정렬
```

# 중요 ********

```python
# 1. 데이터 선택하기 (selection)
# => 원하는 시리즈(칼럼)만 찾는것

df[['A']]
```

```python
df[['A','B']]
```

```python
# 2. 데이터 행 범위 선택하기
df[0:3]
```

```python
# 3. loc => 이름을 사용해서 선택하기
df. loc[:,['A','B']] # ,를 기준으로 앞에는 행(인덱스), 뒤에는 열(칼럼)
```

```python
# 4. iloc => 위치를 사용해서 선택하기

df.iloc[:,0:3]
```

```python
df.iloc[0,2]
```

```python
# 캐글에 있는 데이터를 가져옴
df = pd.read_csv("train.csv")
df
```

```python
tn = df.loc[:,['Pclass','Sex','Age','Survived']]
tn
```

```python
# inplace=True로 설정하면 기존의 값을 변경함 => 불변성을 깸
tn.loc[:,'Sex'].replace(['female','male'],[0,1],inplace=True)
```

```python
tn
```

```python
# 모든 NaN을 정해진 값으로 바꿔줌
# => 일정 칼럼만 바꾸려면 replace()를 사용해야함
# tn.fillna(0)

# NaN이있는 index를날려버림
tn = tn.dropna(axis=0)
tn
```
